```python
# Real-Time Maritime Boat Detection using YOLOv8
## Objectifs
#- Préparer et optimiser le dataset maritime
#- Entraîner un modèle YOLOv8 pour la détection de bateaux
#- Tester le modèle sur des images et vidéos
#- Préparer l’intégration avec StrongSORT (tracking)

## Dataset
#Ships in Aerial Images (Kaggle)
```

```python
from ultralytics import YOLO
import torch
import cv2
import os
import matplotlib.pyplot as plt

print("CUDA available:", torch.cuda.is_available())
```

```python
data_yaml = """
path: C:/Jupyter/Object Tracking/Object Tracking/datasets/maritime_detection

train: train/images
val: valid/images
test: test/images

nc: 1
names: ["boat"]
"""
print(data_yaml)
```

```python
#Vérification & nettoyage des annotations
```

```python
LABEL_PATH = "datasets/maritime_detection/train/labels"

def check_labels(path):
    errors = 0
    for file in os.listdir(path):
        with open(os.path.join(path, file)) as f:
            for line in f:
                values = line.strip().split()
                if len(values) != 5:
                    errors += 1
    print(f"Label errors found: {errors}")

check_labels(LABEL_PATH)
```

```python
#Data Augmentation (Optimisation dataset)
```

```python
import albumentations as A

augmentation = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.Rotate(limit=10, p=0.3),
    A.MotionBlur(p=0.2),
])
```

```python
#Entraînement YOLOv8
```

```python
model = YOLO("yolov8n.pt")

model.train(
    data="datasets/maritime_detection/data.yaml",
    epochs=5,         
    imgsz=416,         
    batch=8,          
    patience=10,
    optimizer="Adam",
    lr0=1e-3,
    name="yolov8_boat"
)

```

```python
#Analyse des résultats
```
